### Dealing with missing numerical data
Mean/median imputation consists of replacing all occurrences of missing values (NA) within a variable by the mean (if the variable has a Gaussian distribution) or median (if the variable has a skewed distribution).

```python
#Filling the missing values of cylinder with the mean value

mean_cyc = cars['Cylinders'].mean()
cars['Cylinders'] = cars['Cylinders'].fillna(mean_cyc)
```

# Catagorical variables

The following catagorical cariables will be transformed.
* Type:   One hot encoding
* Origin: One hot encoding
* Drive Train: One hot encoding

Model:  Unimportant, needs to be dropped

### One Hot Encoding:

One hot encoding, consists of replacing the categorical variable by different boolean variables, which take value 0 or 1, to indicate whether or not a certain category / label of the variable was present for that observation.

Each one of the boolean variables are also known as dummy variables or binary variables.

```python
#One hot encoding

dummy = pd.get_dummies(cars[['Type','Origin', 'DriveTrain']])
dummy.head()
```

```python
dummy.drop(['Type_Hybrid','Origin_Asia', 'DriveTrain_Rear'],axis = 1, inplace = True)
```

```python
dummy.head()
```

```python
#Comboining the dummy with the original data using the concat fuction

cars= pd.concat([cars,dummy], axis = 1)
cars.drop(['Type','Origin', 'DriveTrain','Model'],axis = 1, inplace = True)
cars.head()
```
