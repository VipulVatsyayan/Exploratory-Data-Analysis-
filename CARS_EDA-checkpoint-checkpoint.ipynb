```python
import pandas as pd 
import numpy as np
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
%matplotlib inline 
```

```python
cars = pd.read_csv('C:/Users/vatsy/Desktop/Python/Cars data basics of EDA/CARS.csv')
```

```python
cars.head(5)
```

```python
cars.info()

#  The non-null here does not mean that the columnd dont have a null value, 
#it simply means that a null value in this column is represented as Nan
```

```python
#Checking the number of rows and columns 
cars.shape
```

```python
# Dropping columns  

drop_col = ['MSRP','Invoice']
cars.drop(drop_col, axis=1,inplace=True )
```

```python
# Making sure that the variable is dropped

cars.shape
```

```python
# Taking a look at the column names

cars.columns
```

```python
# Removing duplicates (Here, in this data set we dont have any duplicates)

cars.drop_duplicates(keep='first', inplace=True)
```

```python
#Dropping columns that have the same make

#cars.drop_duplicates(subset= 'Make',keep='first', inplace=True)
```

```python
#cars.head(5)
```

```python
#Checking for missing values

cars.isnull().sum()
```

```python
#Filling the missing values of cylinder with the mean value

mean_cyc = cars['Cylinders'].mean()
cars['Cylinders'] = cars['Cylinders'].fillna(mean_cyc)
```

```python
# Checking for null values

cars.isnull().sum()
```

```python
#sorting the data set using one variable

cars_sort = cars.sort_values(by ='MPG_City',ascending=False)
cars_sort.head()
```

```python
#sorting the data set using two variables

cars_sort = cars.sort_values(by =['MPG_City', 'Cylinders'],ascending= [False, True])
cars_sort.head()
```

```python
#Breif look at the numerical variables
#The describe variable gives the statistics of the data. (only numerical variables)

cars.describe().round(2)
```

Pairplots can also be used for specific variables by mentioning them in the x_vars and Y_vars columns

```python
#Checking for MPG_city variable
sns.pairplot(cars_num,
             kind = 'scatter',
             x_vars= 'MPG_City',
             y_vars=['Weight', 'Horsepower'],
             height=4,
             aspect=1,)
```

```python
#  Using the HUE option
sns.pairplot(cars_num,
             kind = 'scatter',
             x_vars= 'MPG_City',
             y_vars=['Weight', 'Horsepower'],
             height=4,
             aspect=1,
             hue = 'Cylinders')
```
